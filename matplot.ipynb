# multiple lines in one plot

```python
x = [1,2,3,4,5,]
y = [1,22,33,44,55]
z = [10,20,30,40,50]
plt.plot(x,y,'*--',label='X vs Y value')
plt.plot(x,z,'^--',label='X vs Y value')
plt.xlabel('X axis',fontdict={'fontsize':20,'color':'Yellow'})
plt.ylabel('Y axis',fontdict={'fontsize':20,'color':'Green'})
plt.title('Plot betweeen X ,Y and Z',fontdict={'fontsize':20,'color':'Red'})
plt.legend()
```

# generate  1 to 10 x values and generate square values for each number and draw a graph between 1 to 30 value of that vale

```python
import numpy as pd
x = pd.arange(1,11)
y=[i**2 for i in x]
plt.plot(x,y,"*--")
plt.xlabel("X axis",fontdict={'fontsize':20,'color':'Blue'})
plt.ylabel("Y axis",fontdict={'fontsize':20,'color':'Green'})
plt.title("Plot between individual value vs Square values",fontdict={'fontsize':20,'color':'Red'})
```
